# NumPy Installing and importing the library

```python
import numpy as np
```

# Creating NumPy Array

```python
arr=np.array([1,2,3,4,5])
print(arr)
arr[0]=9
print(arr) # Mutable 

arr_2D=np.array([[1,2],[3,4]])
print(arr_2D)
```

# List v/s Array

```python
list=[1,2,3,4,5]
print(list*2)

arr=np.array([1,2,3,4,5]) # 1D vector ki tarah hai yeh
print(arr*2)
```

# Types of array we can make simply
# zeros
# ones
# scalar using full
# identity using eye
# random using random.random
# sequence using (start,stop,step)

```python
zeros=np.zeros((2,2))
print(zeros)

ones=np.ones((2,2))
print(ones)

full=np.full((2,2),9)
print(full)

# Identity Matrix of order n
eye=np.eye(3)
print(eye)

random=np.random.random((2,2))
print(random)

seq=np.arange(0,10,3)
print(seq)
```

# Tensor kya hota hai 3D array

```python
Tensor=np.array([[[1,2],[3,4],[5,6],[7,8]]])
print(Tensor)
```

# Properties of an Array

```python
arr=np.array([[1,8,4],[2,7,0],[3,6,4]])

print("Size:",arr.size) # row * column
print("Shape:",arr.shape)
print("Dimension:",arr.ndim) # 2-D array hi toh hai
print("Dtype:",arr.dtype)
```

# How to reshape an array 
# flatten() v/s ravel()

```python
arr1=np.arange(12)
print(arr1)

arr2=arr1.reshape((3,4))
print(arr2)

arr3=arr2.flatten()
print(arr3)

arr4=arr2.ravel()
print(arr4)
```

# Transpose of an Array

```python
arr=arr2.T
print(arr)
```
