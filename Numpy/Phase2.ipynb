# Importing the library is the 1st step Mandatory 

```python
import numpy as np
```

# Slicing operation just similar to list 

```python
arr=np.array([ [3,2,6], [4,1,5], [5,8,9] ])
print(arr[0,1]) # Specific Element
print(arr[0]) # Entire Row
print(arr[:,1]) # Entire Column
```

# Sorting array wrt row or column

```python
arr1=np.array([3,6,5,1,4,7,9,8,2])
print("Sorted",np.sort(arr1)) # Sorting array

arr2=np.array([ [5,3,4], [2,8,1], [9,7,6] ])
print("Sorted by column", np.sort(arr2, axis = 0)) # Sorting by Column
print("Sorted by row", np.sort(arr2, axis = 1)) # Sorting by Row
```

# Mask/ Condition

```python
arr1=np.array([3,6,5,1,4,7,9,8,2])
even=arr1[ arr1 % 2 == 0 ] # (arr1 % 2 == 0) is mask
print(even)
```

# Similar to above operation

```python
mask = arr1 > 5 
arr2=arr1[mask]
print(arr2)
```

# We are creating a variable which stores indices value as a list

```python
index=np.where( arr1 < 5 )
print(index)
print("array: ",arr1[index])
```

# Condition+Masking

```python
arr=np.array([3,2,7,6,8,4,5,9,1])
condition= np.where(arr>5,"True","False")
print(condition)
```

# Adding v/s Concatenation

```python
arr1=np.array([1,2,3,4])

arr2=np.array([5,6,7,8])
print(arr1+arr2)
arr3=np.concatenate((arr1,arr2))
print(arr3)
```

# Compatibility Test 

```python
a=np.array([1,2,3,4])
b=np.array([3,4,5,6])
c=np.array([5,6,7,8,9])
print("Compatibility",a.shape==b.shape)
print("Compatibility",a.shape==c.shape)
print("Compatibility",c.shape==b.shape)
```

# Adding row or column and deletion of data

```python
a1=np.array([[1,2],[3,4]])

row=np.array([[5,6]]) #we are adding a row

new_a1=np.vstack((a1,row)) #vstack means vertical
print(new_a1)

column=np.array([[7],[8]])

new_a2=np.hstack((a1,column)) #hstack means horizontal
print(new_a2)

arr1=np.array([4,3,6,7,2,8])
delete=np.delete(arr1,2) # element delete karta hai
print(delete)

```
